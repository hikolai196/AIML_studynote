# t-Distributed Stochastic Neighbor Embedding (t-SNE)

t-SNE is a non-linear dimensionality reduction technique that is particularly well-suited for visualizing high-dimensional datasets. It reduces the dimensionality of data while preserving the relative distances between points, making it useful for visualizing clusters.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from sklearn.manifold import TSNE
```

# Load the Iris dataset

```python
iris = load_iris()
X = iris.data
y = iris.target
```

## Standardize the data (mean=0, variance=1)

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

# Perform t-SNE

```python
tsne = TSNE(n_components=2, random_state=42)
X_tsne = tsne.fit_transform(X_scaled)
```

# Plot the t-SNE results

```python
plt.figure(figsize=(8, 6))
markers = ['o', '+', '^']  # 'o' for setosa, '+' for versicolor, '^' for virginica
for target, marker in zip(np.unique(y), markers):
  plt.scatter(X_tsne[y == target, 0], X_tsne[y == target, 1], label=iris.target_names[target], marker=marker)
plt.xlabel('t-SNE Component 1')
plt.ylabel('t-SNE Component 2')
plt.legend()
plt.title('t-SNE of Iris Dataset')
plt.show()
```
