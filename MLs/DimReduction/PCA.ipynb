# Principal Component Analysis (PCA)

Principal Component Analysis (PCA) is a statistical technique used for dimensionality reduction. It transforms the data into a new coordinate system such that the greatest variance by any projection of the data comes to lie on the first coordinate (called the first principal component), the second greatest variance on the second coordinate, and so on. This is particularly useful in reducing the number of variables in a dataset while retaining as much information as possible.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.datasets import load_iris
```

# Load the Iris dataset

```python
iris = load_iris()
X = iris.data
y = iris.target
```

## Standardize the data (mean=0, variance=1)

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

# Perform PCA

```python
pca = PCA(n_components=2)  # Reduce to 2 dimensions
X_pca = pca.fit_transform(X_scaled)
```

# Plot the PCA results

```python
# Define markers for each class
markers = ['o', '+', '^']  # 'o' for setosa, '+' for versicolor, '^' for virginica

# Create subplots
fig, ax = plt.subplots(1, 2, figsize=(16, 6))

# Original Data Plot
for target, marker in zip(np.unique(y), markers):
  ax[0].scatter(X[y == target, 0], X[y == target, 1], label=iris.target_names[target], marker=marker)
ax[0].set_xlabel(iris.feature_names[0])
ax[0].set_ylabel(iris.feature_names[1])
ax[0].legend()
ax[0].set_title('Original Data')

# PCA Transformed Data Plot
for target, marker in zip(np.unique(y), markers):
  ax[1].scatter(X_pca[y == target, 0], X_pca[y == target, 1], label=iris.target_names[target], marker=marker)
ax[1].set_xlabel('Principal Component 1')
ax[1].set_ylabel('Principal Component 2')
ax[1].legend()
ax[1].set_title('PCA of Iris Dataset')

plt.show()
```
