# Multidimensional Scaling (MDS)

MDS is a means of visualizing the level of similarity of individual cases of a dataset. It is a form of non-linear dimensionality reduction.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from sklearn.manifold import MDS
```

# Load the Iris dataset

```python
iris = load_iris()
X = iris.data
y = iris.target
```

## Standardize the data (mean=0, variance=1)

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

# Perform MDS

```python
mds = MDS(n_components=2, random_state=42)
X_mds = mds.fit_transform(X_scaled)
```

# Plot the MDS results

```python
plt.figure(figsize=(8, 6))
markers = ['o', '+', '^']  # 'o' for setosa, '+' for versicolor, '^' for virginica
for target, marker in zip(np.unique(y), markers):
  plt.scatter(X_mds[y == target, 0], X_mds[y == target, 1], label=iris.target_names[target], marker=marker)
plt.xlabel('MDS Component 1')
plt.ylabel('MDS Component 2')
plt.legend()
plt.title('MDS of Iris Dataset')
plt.show()
```
